## Import libraries

```python
import pandas as pd
import numpy as np
from os import walk
```

## Find path of all data files

```python
path = '/home/hakim/Documents/semester 8/DM/HW_3/Data'
file_dic = {}
files_addresses = []
for root, dirs, files in walk(path):
    for file in files:
        if root not in file_dic:
            file_dic[root] = [file]
        else:
            file_dic[root].append(file)
```

```python
data_files = []
for fp in file_dic:
#     data_files_founded = [ f for f in files if f.split('.')[-1] == 'data'] too complex
    for f in file_dic[fp]:
        if f.split('.')[-1] in ['data','trn']:
            full_path = fp + '/' + f
            data_files.append(full_path)
            
for file in data_files:
    print(file)
```

## Open all data file with pandas as dataframe

```python
# df_files = [pd.read_csv(data_file) for data in data_files]
all_df = []
for data_file in data_files:
    all_df.append(pd.read_csv(data_file))
```

```python
for df_indx in range(len(all_df)):
    data_files_name = data_files[df_indx].split('/')[-2::]
    print('\nFile Name --> \t\t', data_files_name[0], ' -> ',data_files_name[1],'\n')
    
    print('\t'*5,'*****Dataset Description*****\n\n',all_df[df_indx].describe())
    
    print('\n','\t'*5,'*****Columns*****\n\n')
    for col in all_df[df_indx].columns: print(col)
    
    print('\n','\t'*5,'*****quantitiy*****\n\n', all_df[df_indx].index)
    df = all_df[df_indx]
    df_cols = df.columns

    for col in df_cols:
        each_value_count = pd.DataFrame(df[col].value_counts())
        
        if len(each_value_count) < 5:
            all_col_weight = []
            print("="*50, col)
            attribute_num = len(each_value_count)
#             make a formul for this
            fair_weight = 100/(attribute_num-1)
            print('attribute num: ', )
            print('Fair Weight: ', fair_weight)
            
            print("Total Records: ", len(df))
            print("Data Type: ", df[col].dtype, end='\n\n')
            print("Missing Value: ", df[col].isnull().values.any(), end='\n\n')
            
            for val in each_value_count.index.values:
                weight = float('{0:.2f}'.format((each_value_count.loc[val][0]/len(df))*100))
                all_col_weight.append(weight)
                print(val, '-----',each_value_count.loc[val][0], '-----', weight,'%')
                
            print('\n')
            
            
            try:
                most_w = max(all_col_weight)
                print("balance attribute: ", (most_w <= fair_weight))
            except:
                print("balance attribute: Unknown")
        
    print('\n\n')
    print("#/"*123)
```
